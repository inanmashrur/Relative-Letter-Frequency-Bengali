```python
import numpy as np
import matplotlib.pyplot as plot
import matplotlib.font_manager as fm
```

```python
Frq = {'অ':0,'আ':0,'ই':0,'ঈ':0, 'উ':0,'ঊ':0,'ঋ':0,'এ':0,'ঐ':0,'ও':0,'ঔ':0,'ক':0,'খ':0,'গ':0,'ঘ':0,'ঙ':0,'চ':0,'ছ':0,'জ':0,'ঝ':0,'ঞ':0,'ট':0,'ঠ':0,'ড':0,'ঢ':0,'ণ':0,'ত':0,'থ':0,'দ':0,'ধ':0,'ন':0,'প':0,'ফ':0,'ব':0,'ভ':0,'ম':0,'য':0,'র':0,'ল':0,'শ':0,'ষ':0,'স':0,'হ':0,'ড়':0,'ঢ়':0,'য়':0,'ৎ':0,'ং':0,'ঃ':0,'ঁ':0}
#len(Frq)
```

```python
def getFrequency(str):
    for char in str:
        if char in Frq:
            Frq[char]+=1  
        else:
            continue
```

```python
IFile = open("read.txt","r",encoding='utf-8')
text = IFile.read()
IFile.close()
```

```python
getFrequency(text)
```

```python
fig, ax = plot.subplots(figsize=(40,23))
prop = fm.FontProperties(fname='Nikosh.ttf')
ax.legend(prop=prop)

font_dirs = ['./', ]
font_files = fm.findSystemFonts(fontpaths=font_dirs)
font_list = fm.createFontList(font_files)
fm.fontManager.ttflist.extend(font_list)

plot.rcParams['font.family'] = 'Nikosh'
plot.tick_params(labelsize=40)

borno_seq = 'অআইঈউঊঋএঐওঔকখগঘঙচছজঝঞটঠডঢণতথদধনপফবভমযরলশষসহড়ঢ়য়ৎংঃঁ'
borno_seq =" ".join(borno_seq.split()) 

freqValues = []
total = 0

for borno in borno_seq:
    count =  Frq[borno]
    freqValues.append(count)
    total += count
    
freqValues = [100*(value/total) for value in freqValues]

plot.bar(range(len(borno_seq)), freqValues, align='center')
plot.xticks(np.arange(len(borno_seq)), list(borno_seq), color='black', rotation=-15)

fig.suptitle('17701068 - Md. Inan Mashrur\nRelative Frequency of Bengali Letters', fontsize=38)

plot.xlabel('অক্ষর-সমূহ', fontsize=40)
plot.ylabel('শতকরা-হার (%)', fontsize=40)

plot.show()
```
